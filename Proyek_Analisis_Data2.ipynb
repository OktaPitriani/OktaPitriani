### Cleaning Data

```python
missing_data = df.isnull().sum()
print(missing_data)
```

```python
df['dteday'] = pd.to_datetime(df['dteday'])
df['year'] = df['dteday'].dt.year
df['month'] = df['dteday'].dt.month
df['weekday'] = df['dteday'].dt.weekday
```

```python
df['cnt'].describe()

df = df[df['cnt'] < df['cnt'].quantile(0.99)]
```

```python
bins = [0, 10, 20, 30, 40]
labels = ['Very Cold', 'Cold', 'Warm', 'Hot']
df.loc[:, 'temp_cat'] = pd.cut(df['temp']*41, bins=bins, labels=labels)
```

```python
avg_rentals_by_day = df.groupby('workingday')['cnt'].mean()
print(avg_rentals_by_day)
```

```python
avg_rentals_by_weather = df.groupby('weathersit')['cnt'].mean()
print(avg_rentals_by_weather)
```
